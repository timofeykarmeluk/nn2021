```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

df = pd.read_csv('2_1nn.csv', sep=';')
df.head()

x = df.iloc[:,:-1]
y = df.iloc[:,-1]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.67, random_state=42)

model = LinearRegression().fit(x_train, y_train)
y_pred = model.predict(x_test)

%matplotlib inline

x_graph = np.arange(len(y_test))

plt.figure(figsize=(20,15))
plt.xlabel("X")
plt.ylabel("Y")
plt.scatter(x_graph, y_test, color='r')
plt.scatter(x_graph, y_pred, color='g')
plt.show()

mean_squared_error(y_test, y_pred)
```
