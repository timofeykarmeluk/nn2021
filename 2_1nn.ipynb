```python
from sklearn.datasets import make_regression
import pandas as pd
import matplotlib.pyplot as plt
x, y = make_regression(
    n_samples=100000,
    n_features=24,
    n_informative=3,
    n_targets=1,
    noise = 0.6,
    bias=10,
    effective_rank = None,
    tail_strength = 0.6,
    shuffle = True,
    coef = False,
    random_state=15)

df = pd.DataFrame(x)
df['target'] = y
df

# df.to_csv('2_1nn.csv', sep=';', index=False)
```
